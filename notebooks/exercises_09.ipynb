#### ðŸ”´ Rectangles

In the follwing list, each tuple contains the length of the sides of a rectangle. Create a file called `areas.csv`, where first line is a header, and each following line represents a rectangle and contains three comma speparated value: length of side 1, length of side 2, area. 

For example, from the list
```
[(10,3),(4,5)]
```
we create a file containing
```
a,b,area
10,3,30
4,5,20
```


````python
rectangles = [
    (3, 7),
    (1, 10),
    (5, 2),
    (8, 4),
    (6, 9)
]

````

<details><summary><u>Solution.</u></summary>
<p>
    
```python
with open('data/areas.csv','w') as out:
    out.write("a,b,area\n")
    for a,b in rectangles:
        out.write(f"{a},{b},{a*b}\n")
```
    
</p>
</details>

#### ðŸ”´ Reprise of Moby Dick

First, download Moby Dick by running the cell below:

````python
import requests # import package

# download resource
response = requests.get("https://posfaim.github.io/dnds5027/data/moby_dick-english.txt") # this returns a special object, not a string
moby_dick = response.text # extract the contents of the webpage into a string
````


The following list contains the names of the main characters from Moby Dick.
* Count how many times they are mentioned in the novel.
* Create a csv file where the first column contains the name and the second contains the count.

<details><summary><u>Hint.</u></summary>

Use the `.count()` string method to get the number of occurances of a substring. For example, `s.count("the")` returns the number of times `"the"` appears in the string `s`.
    
</details>

````python
moby_dick_characters = [
    "Ishmael",
    "Ahab",
    "Queequeg",
    "Starbuck",
    "Stubb",
    "Flask",
    "Boomer",
    "Pip",
    "Fedallah"
]


````

<details><summary><u>Solution.</u></summary>
<p>
    
```python
with open("data/mentions.csv", 'w') as out:
    for character in moby_dick_characters:
        count = moby_dick.count(character)
        out.write(f"{character},{count}\n")
```
    
</p>
</details>
