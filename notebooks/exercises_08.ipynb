#### ðŸ”´ Sorting tuples

Python allows you to sort a list of tuples:

````python
L = [(1,2), (3,4), (3,2), (12,-1), (1,8)]

sorted_L = sorted(L)

sorted_L
````

Can you guess how the tuples are ordered in the sorted list?

<details><summary><u>Solution.</u></summary>
<p>
When you compare two tuples, it first compares the first entry. If the first entry is equal, it compares the second, and so on. You can test this with modifying the following code:
```python
print((1,2)<(3,8))
print((1,1,2)<(1,1,8))
```
So the tuples are sored based on the first number, and ties are resolved using the second.

Tuples provided by `zip`.

````python
birds = ["golden-crowned sparrow", "red-winged blackbird", "whippoorwill", "american gray flycatcher"]
clutch_sizes = [3, 4, 2, 3]

for species num_eggs in zip(birds, clutch_sizes):
    print(f"The {species} typically lays {num_eggs} eggs.")
````

<details><summary><u>Solution.</u></summary>

birds = ["golden-crowned sparrow", "red-winged blackbird", "whippoorwill", "american gray flycatcher"]
clutch_sizes = [3, 4, 2, 3]

for species, num_eggs in zip(birds, clutch_sizes): # add a comma "," between the variables when unpacking a tuple
    print(f"The {species} typically lays {num_eggs} eggs.")

#### ðŸ”´ Returning tuples

Write a function `perimeter_and_area` that takes the length of two sides of a rectangle $a$ and $b$ as arguments, and returns both the perimeter $2a+2b$ and the area $a\cdot b$ of the rectangle. 

<details><summary><u>Solution.</u></summary>
<p>

```python
def perimeter_and_area(a, b):
    perimeter = 2*a+2*b
    area = a*b
    return perimeter, area

print(perimeter_and_area(10, 2))

# or

def perimeter_and_area(a, b):
    return 2*a+2*b, a*b

print(perimeter_and_area(10, 2))
```

#### ðŸ”´ `zip`

We have two lists: one containing driving distances and the other driving durations in the same order. The code calculates the average driving speed for the trips and prints out a warning message if the speed is above the limit of 130 km/hs.

````python
distances = [50, 120, 73, 45] # kms
durations = [40, 100, 11, 20] #minutes

num_trips = len(distances)
for i in range(num_trips):
    speed = distances[i]/ (durations[i]/60) # we need to convert minutes to hours
    if speed>130:
        print(f"Your speed is {speed:.2f} km/h. Slow down!")
````

Re-write the code so that it uses `zip` instead of iterating over possible indices.

<details><summary><u>Solution.</u></summary>
<p>

```python
for dist,dur in zip(distances,durations):
    speed = dist/ (dur/60) # we need to convert minutes to hours
    if speed>130:
        print(f"Your speed is {speed:.2f} km/h. Slow down!")
```
</p>
</details>
