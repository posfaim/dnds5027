#### ðŸ”´ Sorting tuples

Python allows you to sort a list of tuples:

````python
L = [(1,2), (3,4), (3,2), (12,-1), (1,8)]

sorted_L = sorted(L)

sorted_L
````

Can you guess how the tuples are ordered in the sorted list?

<details><summary><u>Solution.</u></summary>
<p>
When you compare two tuples, it first compares the first entry. If the first entry is equal, it compares the second, and so on. You can test this with modifying the following code:
```python
print((1,2)<(3,8))
print((1,1,2)<(1,1,8))
```
So the tuples are sored based on the first number, and ties are resolved using the second.
</p>
</details>

````python
flights_flat = {
    ("New York", "London"): 7.0,
    ("London", "New York"): 7.5,
    ("London", "Paris"): 1.2,
    ("Paris", "London"): 1.3,
    ("Paris", "Rome"): 2.0,
    ("Rome", "Paris"): 2.1,
    ("New York", "Tokyo"): 14.0,
    ("Tokyo", "New York"): 13.5,
    ("Tokyo", "Sydney"): 9.5,
    ("Sydney", "Tokyo"): 9.0,
    ("London", "Rome"): 2.5,
    ("Rome", "London"): 2.4
}

flights_nested = {
    "New York": { "London": 7.0, "Tokyo": 14.0},
    "London":   { "New York": 7.5,"Paris": 1.2, "Rome": 2.5},
    "Paris":    { "London": 1.3, "Rome": 2.0},
    "Rome":     { "Paris": 2.1, "London": 2.4},
    "Tokyo":    {"New York": 13.5,"Sydney": 9.5},
    "Sydney":   {"Tokyo": 9.0}
}
````

Tuples provided by `zip`.

````python
birds = ["golden-crowned sparrow", "red-winged blackbird", "whippoorwill", "american gray flycatcher"]
clutch_sizes = [3, 4, 2, 3]

for species num_eggs in zip(birds, clutch_sizes):
    print(f"The {species} typically lays {num_eggs} eggs.")
````

<details><summary><u>Solution.</u></summary>
<p>
    
```python
birds = ["golden-crowned sparrow", "red-winged blackbird", "whippoorwill", "american gray flycatcher"]
clutch_sizes = [3, 4, 2, 3]

for species, num_eggs in zip(birds, clutch_sizes): # add a comma "," between the variables when unpacking a tuple
    print(f"The {species} typically lays {num_eggs} eggs.")
```
    
</p>
</details>
