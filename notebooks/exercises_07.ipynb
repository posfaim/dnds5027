#### ðŸ”´ Fraction of population

Create a new dictionary that instead of the population of the countries in millions, but the population in terms of percentage of the world population.

<details><summary><u>Hint.</u></summary>

You need a helper variable that keeps track of the total population during summation.
    
</details>

````python
country_populations = {
    "China": 1405.0,
    "India": 1428.0,
    "United States": 342.0,
    "Indonesia": 282.0,
    "Brazil": 216.0,
    "Vatican City": 0.0008,
    "Nauru": 0.012,
    "Tuvalu": 0.011
}

world_pop = 8250.0

````

#### ðŸ”´ Structured data

`weather` is list of dictionaries which contains weather forcast for the next few days: `weather[0]` is tomorrow, `weather[1]` is the day after tomorrow, and so on.

````python
weather = [
    {'min_temp_C': 4, 'max_temp_C': 14, 'percip_mm':0},
    {'min_temp_C': 4, 'max_temp_C': 12, 'percip_mm':1.2},
    {'min_temp_C': -1, 'max_temp_C': 11, 'percip_mm':2.2},
    {'min_temp_C': -2, 'max_temp_C': 8, 'percip_mm':0},
    {'min_temp_C': 3, 'max_temp_C': 16, 'percip_mm':0}
]
````

Write code that prints out tomorrow's maximum temperature.

<details><summary><u>Solution.</u></summary>
<p>
    
```python
    
print(weather[0]['max_temp_C'])

```
    
</p>
</details>

Count the number of days when it rains.

<details><summary><u>Solution.</u></summary>
<p>
    
```python
count = 0
for daily_forecast in weather:
    if daily_forecast['percip_mm']>0:
        count += 1
print(f"The number of rainy days is {count}.")

```
    
</p>
</details>

What is the average minimum temperature in the next three days?


<details><summary><u>Hint.</u></summary>

The next three days is the first three entries in the `weather` list. Use slicing!
</details>

<details><summary><u>Solution.</u></summary>
<p>
    
```python

for daily_forecast in weather[:3]:
    total += daily_forecast['min_temp_C']
```
