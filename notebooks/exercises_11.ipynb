#### ðŸ”´ NumPy -- Rectangles 

The two arrays `side1` and `side2` contain the side lengths of random rectangles. The following code calculates their average area:

````python
side1 = np.random.random(1000)
side2 = np.random.random(1000)

total_area = 0.
for a,b in zip(side1, side2):
    total_area += a*b

avg_area = total_area/len(side1)

print(f"The average area of the rectangles is {avg_area:.2f}")
````

Vectorize the calculation of the average! I.e., calculate the average with array operations and numpy functions without using a `for` loop.

<details><summary><u>Solution.</u></summary>
<p>
    
```python
    
np.mean(side1*side2)
    
```

<details><summary><u>Solution.</u></summary>
