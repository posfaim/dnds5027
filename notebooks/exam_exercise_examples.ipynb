#### üî¥ Max random

Write a function `max_rand` that takes an integer `n` as an argument, generates `n` random floats between 0 and 1 using the `random` module, and returns the maximum number generated.

<details><summary><u>Solution.</u></summary>
<p>
    
```python
    
import random


def max_rand(n):
    rand_nums = []
    for i in range(n):
        rand_nums.append(random.random())
    
    m = 0
    for r in rand_nums:
        if r>m:
            m=r          
    return m

print(max_rand(100))

# or

def max_rand(n):
    m = 0.
    for i in range(n):
        r = random.random()
        if r>m:
            m=r
    return m

print(max_rand(1000))

# or


def max_rand(n):
    rand_nums = []
    for i in range(n):
        rand_nums.append(random.random())
            
    return max(rand_nums)

print(max_rand(100))

    
```

````python
color_hues = {
    "black": (0, 0, 0),
    "white": (255, 255, 255),
    "red": (255, 0, 0),
    "orange": (255, 165, 0),
    "yellow": (255, 255, 0),
    "aquamarine": (127, 255, 212),
    "cerulean": (42, 82, 190),
    "chartreuse": (127, 255, 0),
    "coral": (255, 127, 80),
    "lavender": (230, 230, 250),
    "indigo": (75, 0, 130),
    "crimson": (220, 20, 60),
    "turquoise": (64, 224, 208),
    "sienna": (160, 82, 45),
    "periwinkle": (204, 204, 255),
    "salmon": (250, 128, 114),
    "teal": (0, 128, 128)
}
````

<details><summary><u>Solution.</u></summary>
<p>
    
```python
    
with open("hues.csv","w") as f:
    f.write("name,red,green,blue\n")
    for name, rgb in color_hues.items():
        f.write(f"{name},{rgb[0]},{rgb[1]},{rgb[2]}\n")
```
